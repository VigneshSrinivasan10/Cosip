```python
import tensorflow as tf
import numpy as np

from IPython.display import clear_output, Image, display, HTML
from code_utils import strip_consts, show_graph
```

```python
def neural_network(inputs):
    # Define the operation to be performed
    with tf.variable_scope('XOR'):
        # Define the weight variables for each layer
        w1 = tf.Variable(tf.random_uniform([2,2], -1, 1), name='w1')
        w2 = tf.Variable(tf.random_uniform([2,1], -1, 1), name='w2')
        b1 = tf.Variable(tf.zeros([2]), name = 'b1')
        b2 = tf.Variable(tf.zeros([1]), name = 'b2')
                    
        # Create the operations to be executed 
    
        # intermediate = ab * w1 + b1
        with tf.variable_scope('hidden_layer'):
            hidden_layer = tf.matmul(inputs, w1) + b1
            hidden_layer = tf.nn.tanh(hidden_layer)
        # output = intermediate * w2 + b2
        with tf.variable_scope('output'):
            final_layer = tf.matmul(hidden_layer, w2, name='output') + b2
            output = tf.nn.sigmoid(final_layer) 
        return final_layer#output
```

```python
''' Tensorflow codes - XOR'''

# XOR 
# 00 - 0
# 01 - 1
# 10 - 1
# 11 - 0

tf.reset_default_graph()
# Define the inputs 
with tf.variable_scope('inputs/ab'):
    ab = tf.placeholder(tf.float32, shape=[4,2])
with tf.variable_scope('ground_truth'):
    ground_truth= tf.placeholder(tf.float32, shape=[4,1])




output = neural_network(ab)
#print(ab,hidden_layer,output)

# Define loss function and optimizers
learning_rate = 0.10
with tf.variable_scope('mean_squared_error'):
#     loss = tf.nn.softmax_cross_entropy_with_logits(labels=ground_truth, logits=output)
#     loss = tf.reduce_mean(loss)
     loss = tf.losses.mean_squared_error(ground_truth, output)
#optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(loss)
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(loss)
```

```python
#Create a tensorflow session
with tf.Session() as sess:
    # initialize all the variables - this assigns the values to the variables
    tf.global_variables_initializer().run()
    
    # Create the XOR table
    input_ = np.array([[0,0],[0,1],[1,0],[1,1]]).reshape([4,2])
    gt_ = np.array([0,1,1,0]).reshape([4,1])
    
    # The dictionary assigns the numerical values to the placeholders
    values_dict = {ab:input_, ground_truth:gt_}
    
    max_steps = 1000
    for i in range(max_steps):
        # sess.run() consists of 'fetch' and 'feed' 
        # 'fetch' are the list of variables to be returned
        # 'feed' is the dictionary of values assigned to respective placeholders
        _, input1, output_, loss_ = sess.run([optimizer, ab,output, loss], feed_dict=values_dict)
    print('Trained')
    
```
